```python
from huggingface_hub import list_models
from datetime import datetime
import pandas as pd
import json


# Use root method


def get_model():
    print("Getting models from Huggingface: downloading...")

    current_datetime = datetime.now()
    formatted_datetime = current_datetime.strftime("%Y-%m-%d %H:%M:%S")
    models = list_models()

    model_lists = list(models)
    print("Getting models from Huggingface: successfully getting models")
    print(f"There are {len(model_lists)} models as of {formatted_datetime}")
    print()
    return model_lists , formatted_datetime


def add_attributes(model_lists, source_name, formatted_datetime):
    print("Adding source and name time attributes: adding...")
    for item in range(len(model_lists)):
        model_lists[item].__dict__['source'] = source_name
        model_lists[item].__dict__['download_datetime'] = formatted_datetime
    
    print("Adding source and name time attributes: successfully adding attributes")

    return model_lists


def save_to_json(model_lists):
    print()
    print('Saving file to json...')
    current_date = datetime.now().strftime("%Y%m%d")
    # Create the file name with the date
    file_name = f"{current_date}_model_list.json"
    # Save the list as JSON with the generated file name
    with open(file_name, 'w') as file:
        json.dump([obj.__dict__ for obj in model_lists], file)


def save_to_csv(model_lists):
    print()
    print('Saving file to csv...')
    current_date = datetime.now().strftime("%Y%m%d")
    # Create the file name with the date
    file_name = f"{current_date}_model_list.csv"
    model_df = pd.DataFrame(d.__dict__ for d in model_lists )
    model_df.to_csv(file_name)



if __name__ == "__main__":
    # This code will only run when the script is executed directly
    # It will not run when the script is imported as a module
    model_lists , formatted_datetime = get_model()
    model_lists = add_attributes(model_lists, 
                                 source_name='HuggingFace', 
                                 formatted_datetime = formatted_datetime)
    save_to_json(model_lists)
    save_to_csv(model_lists)
```
