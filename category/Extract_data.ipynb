```python
import re
from huggingface_hub import list_models

models = list_models()
model_lists = list(models)

load_local = [] #Download column
quantize_model = [] #Quantize column

"""
download_dict = {1 : "Download Version",
                 0 : "Call API Version"}

quantize_dict = {1 : "Quantize Version",
                 0 : "Not Quantize Version"}
"""

def check_download(id):
    if re.search(r'GGML|GGUF', id, re.IGNORECASE):
        return 1 #Download Version
    else :
        return 0 #Call API Version
def check_quantized(id):
    if re.search(r'GGML|GGUF|8bit|4bit|2bit', id, re.IGNORECASE):
        return 1 #Quantize Version
    else :
        return 0 #Not Quantize Version

for model in model_lists:
    model_name = model.id.lower()
    load_local.append(check_download(model_name))
    quantize_model.append(check_quantized(model_name))
```
